```python
# Import libraries for data analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in the data from the csv file
data_checking = pd.read_csv('checkin_checkout_history_updated.csv')

# Print the first 5 rows of the data
data_checking.head()
```

```python
# Read in the data from the csv file
data_gym = pd.read_csv('gym_locations_data.csv')

# Print the first 5 rows of the data
data_gym.head()
```

```python
# Read in the data from the csv file
data_sub = pd.read_csv('subscription_plans.csv')

# Print the first 5 rows of the data
data_sub.head()
```

```python
# Read in the data from the csv file
data_users = pd.read_csv('users_data.csv')

# Print the first 5 rows of the data
data_users.head()
```

```python
# Merge data_checking with data_users on 'user_id'
data_merged = pd.merge(data_checking, data_users, on='user_id')

# Merge the result with data_gym on 'gym_id'
data_merged = pd.merge(data_merged, data_gym, on='gym_id')

# Merge the result with data_sub on 'subscription_plan'
data_final = pd.merge(data_merged, data_sub, on='subscription_plan')

# Display the information of the resulting dataframe
data_final.info()

# Display the first 5 rows of the resulting dataframe
data_final.head()
```

```python
# Convert 'checkout_time' and 'checkin_time' columns to datetime
data_final['checkout_time'] = pd.to_datetime(data_final['checkout_time'])
data_final['checkin_time'] = pd.to_datetime(data_final['checkin_time'])

# Add a new column "time spent" to the dataframe
data_final['time spent'] = data_final['checkout_time'] - data_final['checkin_time']

# Format "time spent" to show only hours, minutes, and seconds
data_final['time spent'] = data_final['time spent'].apply(lambda x: str(x).split(' ')[-1])


# Display the first 5 rows of the resulting dataframe
data_final.head()

# Display the information of the resulting dataframe
data_final.info()
```

```python
# Create a new dataframe with the range 50-64 years old for analysys
data_final_50_64 = data_final[(data_final['age'] >= 50) & (data_final['age'] <= 64)]

# Display the first 5 rows of the resulting dataframe
data_final_50_64.head()

# Display the information of the resulting dataframe
data_final_50_64.info()
```

```python
# Apply a style to the plot
sns.set_style("whitegrid")
```

```python
# Plot the subscription plan distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data_final_50_64, x='subscription_plan', palette='viridis')
plt.title('Subscription Plan Distribution (Age 50-64)')
plt.xlabel('Subscription Plan')
plt.ylabel('Count')
plt.show()
```

```python
# Group by subscription plan and count the number of users
subscription_counts = data_final_50_64.groupby('subscription_plan').size().reset_index(name='count')

# Plot pie chart for the subscription plan distribution
plt.figure(figsize=(8, 8))
plt.pie(subscription_counts['count'], labels=subscription_counts['subscription_plan'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(subscription_counts)))
plt.title('Distribucin del Plan de Suscripcin')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Group by subscription plan and count the number of users
subscription_counts = data_final_50_64.groupby('subscription_plan').size().reset_index(name='count')

# Display the result
print(subscription_counts)
```

```python
# Plot the workout type distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data_final_50_64, x='workout_type', palette='viridis')
plt.title('Workout Type Distribution (Age 50-64)')
plt.xlabel('Workout Type')
plt.ylabel('Count')
plt.show()
```

```python
# Create 'time spent (minutes)' column if it doesn't exist
if 'time spent (minutes)' not in data_final_50_64.columns:
    data_final_50_64['time spent (minutes)'] = data_final_50_64['time spent'].apply(lambda x: pd.to_timedelta(x).total_seconds() / 60)

# Create a new column with time ranges in minutes
bins = [0, 30, 60, 90, 120, 150, 180]
labels = ['0-30', '30-60', '60-90', '90-120', '120-150', '150-180']
data_final_50_64['time_bins'] = pd.cut(data_final_50_64['time spent (minutes)'], bins=bins, labels=labels, right=False)

# Plot the distribution of time spent in the gym by time ranges
plt.figure(figsize=(10, 6))
sns.countplot(data=data_final_50_64, x='time_bins', palette='viridis')
plt.title('Distribucin del Tiempo Pasado en el Gimnasio (Edad 50-64)')
plt.xlabel('Tiempo Pasado (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Plot the gym type distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data_final_50_64, x='gym_type', palette='viridis')
plt.title('Gym Type Distribution (Age 50-64)')
plt.xlabel('Gym Type')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of calories burned by workout type
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_final_50_64, x='workout_type', y='calories_burned', palette='viridis')
plt.title('Calories Burned by Workout Type (Age 50-64)')
plt.xlabel('Workout Type')
plt.ylabel('Calories Burned')
plt.show()
```

```python
# Distribution of users by gender and subscription plan
plt.figure(figsize=(10, 6))
sns.countplot(data=data_final_50_64, x='subscription_plan', hue='gender', palette='viridis')
plt.title('Users by Gender and Subscription Plan (Age 50-64)')
plt.xlabel('Subscription Plan')
plt.ylabel('Count')
plt.show()
```

```python
# Frequency of gym visits by user location
plt.figure(figsize=(10, 6))
sns.countplot(data=data_final_50_64, x='user_location', palette='viridis')
plt.title('Frequency of Gym Visits by User Location (Age 50-64)')
plt.xlabel('User Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average time spent in gym by gym type for users aged 50-64
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_final_50_64, x='gym_type', y='time spent (minutes)', palette='viridis')
plt.title('Average Time Spent in Gym by Gym Type (Age 50-64)')
plt.xlabel('Gym Type')
plt.ylabel('Time Spent (minutes)')
plt.show()
```

```python
# Convert 'sign_up_date' and 'checkout_time' to datetime if they are not already
data_final_50_64['sign_up_date'] = pd.to_datetime(data_final_50_64['sign_up_date'])
data_final_50_64['checkout_time'] = pd.to_datetime(data_final_50_64['checkout_time'])

# Calculate the 'membership_duration' in days
data_final_50_64['membership_duration'] = (data_final_50_64['checkout_time'] - data_final_50_64['sign_up_date']).dt.days

# Plot the membership duration by gym type
plt.figure(figsize=(10, 6))
sns.violinplot(data=data_final_50_64, x='gym_type', y='membership_duration', palette='viridis')
plt.title('Membership Duration by Gym Type (Age 50-64)')
plt.xlabel('Gym Type')
plt.ylabel('Membership Duration (days)')
plt.show()
```
